```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
df = pd.read_csv("https://github.com/cheejiayuan512/data/raw/master/order_brush_order.csv", index_col = "shopid")
```

```python
df['event_time'] = df.apply(lambda x: dt.datetime.strptime(x.event_time,  "%Y-%m-%d %H:%M:%S" ), axis = 1)
```

```python
df = df.groupby(['shopid']).agg(lambda x: tuple(x)).applymap(list).reset_index()
```

```python
df['zipped'] = df.apply(lambda x: list(zip(x.event_time,x.userid)), axis=1)
```

```python
df['zipped']= df['zipped'].apply(sorted, reverse= True)
df
```
